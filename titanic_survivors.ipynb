```python
train_big = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
age_slice = train_big['Age'].copy()
age_slice[np.isnan(age_slice)] = train_big['Age'].mean()
train_big['Age'] = age_slice
train_big.isnull().sum()
```

```python
genders = {'female':1, 'male':0}
train_big["Sex"] = train_big["Sex"].map(genders)
# Filling 2 missing values with most frequent value
train_big["Embarked"] = train_big["Embarked"].fillna('S')
```

```python
ports = {"S":0, "C":1, "Q":2}
train_big["Embarked"] = train_big["Embarked"].map(ports)
```

```python
train_big["Family"] = train_big["SibSp"] + train_big["Parch"]
```

```python
df_train_x = train_big[['Pclass','Sex', 'Age', 'Family', 'Embarked']]
df_train_y = train_big[['Survived']]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df_train_x = ss.fit_transform(df_train_x)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_train_x, df_train_y, test_size=0.20, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree,svm
from sklearn.metrics import accuracy_score
```

```python
rfg = RandomForestClassifier()
rfg.fit(train_x, train_y)
rfg_pred = rfg.predict(test_x)
print(accuracy_score(test_y, rfg_pred))
```

```python
lg = LogisticRegression()
lg.fit(train_x, train_y)
lg_pred = lg.predict(test_x)
print(accuracy_score(test_y, lg_pred))
```

```python
from sklearn.svm import SVC

clf = SVC(kernel='linear')
clf.fit(train_x,train_y)

y_pred = clf.predict(test_x)
print(accuracy_score(test_y,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(train_x, train_y)

y_pred = nb.predict(test_x)
print(accuracy_score(test_y, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
dt.fit(train_x, train_y )
y_pred = dt.predict(test_x)
print(accuracy_score(test_y, y_pred))
```
